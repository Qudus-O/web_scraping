```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
response = requests.get("https://news.ycombinator.com/news")
soup = BeautifulSoup(response.text, 'html.parser')

article_tags = soup.find_all("span", class_= "titleline")
article_upvotes = soup.find_all("span", class_= "score")
```

```python
article_text = [tag.get_text().strip() for tag in article_tags]
article_links = [tag.a.get("href") for tag in article_tags if tag.a]
article_upvotes = [int(score.getText().split()[0]) for score in soup.find_all(name="span", class_="score")]

print(article_text)
print(article_links)
print(article_upvotes)
```

```python
article_df = pd.DataFrame(list(zip(article_text, article_links, article_upvotes)), columns=["Articles", "Links", "Upvotes_Points"])
article_df.head()
```
